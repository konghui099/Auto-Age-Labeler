```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('audio_training_data_cleaned.csv')
df.columns
```

```python
df = df.drop(columns=['Unnamed: 0','filename'])
```

```python
sns.set(style='darkgrid')
plt.figure(figsize=(8,4))
sns.countplot(x='age',data=df,color='b')
```

```python
f,axes = plt.subplots(2,4,figsize=(20,8))

sns.distplot(df[df['age']=='teens']['mean energy'],ax=axes[0,0])
sns.distplot(df[df['age']=='twenties']['mean energy'],ax=axes[0,1])
sns.distplot(df[df['age']=='thirties']['mean energy'],ax=axes[0,2])
sns.distplot(df[df['age']=='fourties']['mean energy'],ax=axes[0,3])

sns.distplot(df[df['age']=='fifties']['mean energy'],ax=axes[1,0])
sns.distplot(df[df['age']=='sixties']['mean energy'],ax=axes[1,1])
sns.distplot(df[df['age']=='seventies']['mean energy'],ax=axes[1,2])
sns.distplot(df[df['age']=='eighties']['mean energy'],ax=axes[1,3])

axes[0,0].set_xlabel('Mean Energy (teens)')
axes[0,1].set_xlabel('Mean Energy (twenties)')
axes[0,2].set_xlabel('Mean Energy (thirties)')
axes[0,3].set_xlabel('Mean Energy (fourties)')
axes[1,0].set_xlabel('Mean Energy (fifties)')
axes[1,1].set_xlabel('Mean Energy (sixties)')
axes[1,2].set_xlabel('Mean Energy (seventies)')
axes[1,3].set_xlabel('Mean Energy (eighties)')
```

```python
plt.figure(figsize=(4,8))
ax = sns.boxplot(x="age",y="mean energy",data=df)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
plt.ylim(top=0.05)
```

```python
f,axes = plt.subplots(2,4,figsize=(20,8))

sns.distplot(df[df['age']=='teens']['mean entropy of energy'],ax=axes[0,0])
sns.distplot(df[df['age']=='twenties']['mean entropy of energy'],ax=axes[0,1])
sns.distplot(df[df['age']=='thirties']['mean entropy of energy'],ax=axes[0,2])
sns.distplot(df[df['age']=='fourties']['mean entropy of energy'],ax=axes[0,3])

sns.distplot(df[df['age']=='fifties']['mean entropy of energy'],ax=axes[1,0])
sns.distplot(df[df['age']=='sixties']['mean entropy of energy'],ax=axes[1,1])
sns.distplot(df[df['age']=='seventies']['mean entropy of energy'],ax=axes[1,2])
sns.distplot(df[df['age']=='eighties']['mean entropy of energy'],ax=axes[1,3])

axes[0,0].set_xlabel('Mean Entropy of Energy (teens)')
axes[0,1].set_xlabel('Mean Entropy of Energy (twenties)')
axes[0,2].set_xlabel('Mean Entropy of Energy (thirties)')
axes[0,3].set_xlabel('Mean Entropy of Energy (fourties)')
axes[1,0].set_xlabel('Mean Entropy of Energy (fifties)')
axes[1,1].set_xlabel('Mean Entropy of Energy (sixties)')
axes[1,2].set_xlabel('Mean Entropy of Energy (seventies)')
axes[1,3].set_xlabel('Mean Entropy of Energy (eighties)')
```
